## Operador Roberts
O operador Roberts é um operador de gradiente discreto usado para detectar bordas em uma imagem. Ele calcula a magnitude do gradiente da imagem usando duas máscaras de tamanho 2x2. O operador Roberts tem duas máscaras - uma para detecção de bordas horizontais e outra para bordas verticais. O código 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# carrega a imagem em escala de cinza
img = cv2.imread('Draven.jpg', 0)

# aplica o operador de Roberts para a detecção de bordas na direção horizontal
roberts_x = np.array([[1, 0], [0, -1]], dtype=np.float32)
img_x = cv2.filter2D(img, -1, roberts_x)

# aplica o operador de Roberts para a detecção de bordas na direção vertical
roberts_y = np.array([[0, 1], [-1, 0]], dtype=np.float32)
img_y = cv2.filter2D(img, -1, roberts_y)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))


# calcula a magnitude do gradiente
magnitude = np.sqrt(img_x**2 + img_y**2)

# plota as imagens resultantes
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title('Imagem original'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(img_x, cmap='gray')
plt.title('Gradiente na direção horizontal'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(img_y, cmap='gray')
plt.title('Gradiente na direção vertical'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(magnitude, cmap='gray')
plt.title('Magnitude do gradiente'), plt.xticks([]), plt.yticks([])

plt.show()

```

## Operador Prewitt

O operador Prewitt é um operador de gradiente discreto que também é usado para detectar bordas em uma imagem. Ele calcula a magnitude do gradiente usando duas máscaras de tamanho 3x3. O operador Prewitt tem duas máscaras - uma para detecção de bordas horizontais e outra para bordas verticais. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# carrega a imagem em escala de cinza
img = cv2.imread('Draven.jpg', 0)

# aplica o operador de Prewitt para a detecção de bordas na direção horizontal
prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32)
img_x = cv2.filter2D(img, -1, prewitt_x)

# aplica o operador de Prewitt para a detecção de bordas na direção vertical
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=np.float32)
img_y = cv2.filter2D(img, -1, prewitt_y)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))


# calcula a magnitude do gradiente
magnitude = np.sqrt(img_x**2 + img_y**2)

# plota as imagens resultantes
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title('Imagem original'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(img_x, cmap='gray')
plt.title('Gradiente na direção horizontal'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(img_y, cmap='gray')
plt.title('Gradiente na direção vertical'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(magnitude, cmap='gray')
plt.title('Magnitude do gradiente'), plt.xticks([]), plt.yticks([])

plt.show()
```

## Operador Sobel

O operador Sobel é outro operador de gradiente discreto que é usado para detectar bordas em uma imagem. Ele usa duas máscaras de tamanho 3x3 para calcular a magnitude do gradiente. O operador Sobel tem duas máscaras - uma para detecção de bordas horizontais e outra para bordas verticais. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Lendo a imagem
img = cv2.imread('Draven.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicando o operador de gradiente Sobel
grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Calculando a magnitude do gradiente
grad_mag = np.sqrt(grad_x**2 + grad_y**2)

# Plotando as imagens
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('Original')

axs[0, 1].imshow(grad_x, cmap='gray')
axs[0, 1].set_title('Gradiente horizontal')

axs[1, 0].imshow(grad_y, cmap='gray')
axs[1, 0].set_title('Gradiente vertical')

axs[1, 1].imshow(grad_mag, cmap='gray')
axs[1, 1].set_title('Magnitude do gradiente')

for ax in axs.flat:
    ax.set(xlabel='Pixels', ylabel='Pixels')

plt.show()
```
